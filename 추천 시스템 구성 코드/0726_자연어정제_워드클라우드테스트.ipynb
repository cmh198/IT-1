```python
import pandas as pd
pd.set_option('display.max_row', 4000)
pd.set_option('display.max_columns', 100)
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
from konlpy.tag import Okt # 옛날 Twitter 클래스
from konlpy.tag import Kkma
import re

okt = Okt()
kkma = Kkma()
```

```python
# 다음리뷰

blog = pd.read_csv("블로그최종.csv")
blog_ = blog.loc[:,['순번','영화명','content']]
blogwords = blog_['content'][1:2].sum()
```

```python
blogwords
```

```python
# 정규표현식
letters_only = re.sub('[^0-9가-힣\s\.\!\?]*', '', blogwords)
letters_only= letters_only.replace('\n', '')
# letters_only = re.sub('[\n]*', '', letters_only) # \n 없애는 코드
# letters_only = re.sub('[\s]*', '', letters_only) # 띄어쓰기 다 붙이는 코드
letters_only
```

```python
okt = Okt()
# 정제된 리뷰로 형태소 테이블 만들기
sampledf = pd.DataFrame(okt.pos(letters_only, norm=True, stem=True))
# 명사, 형용사, 동사만 추출하기
sampledf = sampledf[(sampledf[1]=='Noun')|(sampledf[1]=='Adjective')|(sampledf[1]=='Verb')]
# 2글자 이상인 단어만 가져와서 리스트 만들기
spword = []
for word in sampledf[0]:
    if len(word)>=2:
        spword.append(word)
# 단어별 등장횟수 세기
pd.DataFrame(spword)[0].unique()
# 조사 / 한글자단어 빠진 상태이기 때문에 최대한 정제된 상황임
# 이때 나오는 불용어 (예 : 어떻다) 등을 제거하기 위한 우리만의 불용어사전만들기 (플랫폼별로 하기) !
```

```python
stopwords2 = pd.read_csv('stopwords726_5.txt') # 데이터프레임으로 불러옴
```

```python
stopwords2 = stopwords2.astype('str') # 혹시몰라서 str타입으로 바꿈
```

```python
stopwords2 = np.array(stopwords2) # np.array로 바꿔줘야 적용됨
```

```python
okt = Okt()
# 정제된 리뷰로 형태소 테이블 만들기
sampledf = pd.DataFrame(okt.pos(letters_only, norm=True, stem=True))
# 명사, 형용사, 동사만 추출하기
sampledf = sampledf[(sampledf[1]=='Noun')|(sampledf[1]=='Adjective')|(sampledf[1]=='Verb')]
# 2글자 이상인 단어만 가져와서 리스트 만들기
spword = []
for word in sampledf[0]:
    if len(word)>=2:
        if word not in stopwords2: # 스탑워즈 제외하고 출력하기
            spword.append(word)
# 단어별 등장횟수 세기
pd.DataFrame(spword)[0].unique()
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image

text = spword
lion_mark = np.array(Image.open('18855153-영화-로고.jpg'))
```

```python
import collections
textdict = collections.Counter(text) # 리스트에 있는 text의 단어 빈도수
```

```python
import matplotlib.pyplot as plt
import platform

import matplotlib
from matplotlib import font_manager, rc
if platform.system()=="Windows":
    font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/ALONG.ttf").get_name()
    rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False
import warnings
warnings.filterwarnings("ignore")
    
%matplotlib inline
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(lion_mark, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
a = WordCloud(background_color='white', max_words=2000, mask=lion_mark,font_path='c:/Windows/Fonts/ALONG.ttf').generate_from_frequencies(textdict)
# 딕셔너리 타입을 워드클라우드에 넣기(비율대로)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(a, interpolation='bilinear')
plt.axis('off')
plt.show()
# 뽑음
```

```python
blog = pd.read_csv('../FinalProject/movieData_word/naverBlog_word_rev.csv')
blog.drop('content', axis=1, inplace=True)
blog = blog[['영화명', 'word']]
blog.head()
```

```python
daum = pd.read_csv('../FinalProject/movieData_word/daumReview_word_rev.csv')
daum = daum[['영화명', 'word']]
daum.head()
```

```python
watcha = pd.read_csv('../FinalProject/movieData_word/watcha_word0728.csv')
watcha = watcha[['영화명', 'word']]
watcha.head()
```

```python
naver = pd.read_csv('../FinalProject/movieData_word/naver.csv')
naver = naver.rename(columns={'name':'영화명'})
naver.head()
```

```python
movie = pd.read_excel('../FinalProject/movieData_word/moviedata2.xlsx')
movie.head()
```

```python
merge_1 = movie.merge(watcha, how='left', on='영화명')
```

```python
merge_1.head()
```

```python
merge_2 = merge_1.merge(daum, how='left', on='영화명')
```

```python
merge_2.head()
```

```python
merge_3 = merge_2.merge(blog, how='left', on = '영화명')
```

```python
merge_3.head()
```

```python
# 김진현 보류
merge_4 = merge_3.merge(naver, how='left', on ='영화명')
```

```python
merge_4.head()
```

```python
merge_4.columns = ['순번', '영화명', '감독', '개봉일', '국적', '전국관객수', '장르', '등급', 'word_x', 'word_y', 'word_z', 'word_a']
```

```python
merge_4['word_x'] = merge_4['word_x'].fillna("")
merge_4['word_y'] = merge_4['word_y'].fillna("")
merge_4['word_z'] = merge_4['word_z'].fillna("")
merge_4['word_a'] = merge_4['word_a'].fillna("")
```

```python
merge_4['word'] = merge_4['word_x'] + merge_4['word_y'] + merge_4['word_z'] + merge_4['word_a']
```

```python
merge_4.drop(['word_x', 'word_y', 'word_z', 'word_a'], axis=1, inplace=True)
```

```python
merge_4['word'] = merge_4.word.str.replace('[^0-9|가-핳| ]', '')
```

```python
merge_4.head()
```

```python
merge_4 = merge_4[merge_4['word'] != ""]
```

```python
merge_4 = merge_4.drop_duplicates('영화명', keep = 'first')
```

```python
merge_4.to_csv('final_movie_1.csv', encoding='utf-8-sig', index=False)
```

# 위까지 merge 완료, 아래부터 단어 분리

```python
movie = pd.read_csv('../FinalProject/movieData_word/final_movie_1.csv')
```

```python
# word에 있는 단어들 다 분리해서 word2에 넣기
movie['word2'] = 0
for i in range(len(movie['word'])) :
    movie['word2'][i] = re.sub('[^가-힣\s]*', '',str(movie['word'][i])).split(" ")
```

```python
# word2가 리스트니까 단어 다 합쳐버리기
word2 = []
for i in movie['word2'] :
    for j in range(len(i)) :
        word2.append(i[j])
```

# pickle 사용해서 영화명 : word 딕셔너리 만들기

```python
import pickle
```

```python
# 쪼갠 단어로 딕셔너리
# movie_dict = {}
# for i in range(0, len(movie['word2'])) :
#     movie_dict[movie['영화명'][i]] = movie['word2'][i]
```

```python
# 단어 이어 붙인거로 딕셔너리 만들기
movie_dict = {}
for i in range(0, len(movie['word'])) :
    movie_dict[movie['영화명'][i]] = movie['word'][i]
```

```python
movie_dict
```

```python
# save
with open('movie_dict.pickle', 'wb') as f:
    pickle.dump(movie_dict, f, pickle.HIGHEST_PROTOCOL)
```

# 다음 한줄평으로 영화명 : [리뷰들] 딕셔너리 만들기

```python
# 다음리뷰

daum = pd.read_csv("정빵무비.csv")
daum_ = daum.loc[:,['영화명','reviews']]
daum_.head()
```

```python
a = daum_['reviews'][7]
a
```

```python
chat = re.findall("\\\'([가-힣0-9a-zA-Z\.\s]+)\\\'",a)
chat
```

```python
# 영화명 : [리뷰들] 딕셔너리 만들기
chat_dict = {}
for i in range(len(daum_)):
    a = daum_['reviews'][i]
    chat = re.findall("\\\'([가-힣0-9a-zA-Z\s]+)\\\'",a)
    chat_dict[daum_['영화명'][i]] = chat 
```

```python
chat_dict
```

```python
# 피클타입으로 저장
with open('chat_dict.pickle', 'wb') as f:
    pickle.dump(chat_dict, f, pickle.HIGHEST_PROTOCOL)
```
